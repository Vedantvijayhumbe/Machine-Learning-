```python
import matplotlib.pyplot as plt
import seaborn as sns

# For example, to plot a histogram of a specific column (say 'column_name')
plt.figure(figsize=(10, 6))
sns.histplot(df['Study Hours'], kde=True)  # histogram with a kernel density estimate
plt.title('Histogram of Study Hours')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

this is just to compare the results of profile report library and matplotlib library .

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot using seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Study Hours'], y=df['Grades'])
plt.title('Scatter Plot:  Study Hours vs Grades')
plt.xlabel('Study Hours')
plt.ylabel('Grades')
plt.show()
```

here we can see that most increase in the number of hours of studies also contributes to higher grades . also we can observe that there's a particular threshold hours most students need to cross in order to maximize their return on investment of hours they study .
here we can see some outliers too , but overall graph emphasises on importance of hardwork . ✅

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['Study Hours'], y=df['Grades'])
plt.title('Line Plot: Study Hours vs Grades')
plt.xlabel('Study Hours')
plt.ylabel('Grades')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df['Study Hours'], y=df['Grades'])
plt.title('Bar Plot: Study Hours vs Grades')
plt.xlabel('Study Hours ')
plt.ylabel('Grades')
plt.show()
```

```python
sns.pairplot(df[['Study Hours', 'Grades']])
plt.show()
```

```python
corr_matrix = df[['Study Hours', 'Attendance (%)' ,'Sleep Hours' , 'Socioeconomic Score' , 'Grades' ]].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap: Study Hours vs Grades')
plt.show()
```
